```
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import os
```

Question 3C

```
# get theta vals
theta_hat_vals = np.linspace(0.0, 1.0, num=101)

# get data points
n = 10
n_ones = 6
ones = np.array([1]*n_ones)
zeros = np.array([0]*(n-n_ones))
all_data = np.concatenate((ones,zeros))

# get log likelihood for each data point
def log_likelihood(theta_val, data):
  result = 1
  for point in data:
    result *= (theta_val ** point) * ((1-theta_val) ** (1-point))
  
  return result

# get log likelihood val for each theta val
y = np.array([])

for theta in theta_hat_vals:
  y = np.append(y, log_likelihood(theta, all_data))

MLE = theta_hat_vals[np.argmax(y)]

plt.xlabel('theta value')
plt.ylabel('Likelihood')
plt.title('Likelihood for Size-10 Dataset')
plt.plot(theta_hat_vals, y)
axes = plt.gca()
axes.set_ylim([-.0001,0.0013])
plt.plot(MLE,-.0001,'ro', markersize=12) 

plt.savefig("/content/gdrive/My Drive/schoolwork/3rd_year/winter/cs_m146/hw_2/q3c.png")
```
